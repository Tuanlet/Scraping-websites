# making dynamic scraping for multiple pages


```python
# we need to make link dynamic so it searches not just page 1 of the website

data = []

for i in tqdm(range(1,51)):                                     #range in 1- 51 for number of pages on website, change to tqdm for process bar
    #print(f"Scraping page {i}")
    link = 'https://books.toscrape.com/catalogue/page-' + str(i) +'.html' #creates link for each page
    
    res  = requests.get(link)                                     # Sending Request to the link
    #items= soup = BeautifulSoup(res.text, 'html.parser')     #THIS WAS DEBUGGING TEST           
    soup = BeautifulSoup(res.text, 'html.parser')                  # Creating a soup for that page
    print(f"Found {len(items)} items on page {i}")
    
    for sp in soup.find_all('li', class_ ="col-xs-6 col-sm-4 col-md-3 col-lg-3"):

    
        img_link   = 'https://books.toscrape.com/' + sp.find('img').get('src')[3:]
        book_link  = 'https://books.toscrape.com/catalogue/' +sp.find_all('a')[-1].get('href')    
        title      = sp.find_all('a')[-1].get('title')
        rating     = sp.find('p').get('class')[-1]
        stock      = sp.find('p', class_ ='instock availability').text.strip()
        price      = sp.find('p', class_ = 'price_color').text[1:]

    
        data.append([title,rating,price,stock,book_link,img_link])  #saves crapped data in a list
#print(f"Total items scraped: {len(data)}")
```
