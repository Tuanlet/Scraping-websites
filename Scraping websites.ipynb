```python
data = []

for sp in soup.find_all('div', class_ = 'quote'): #seach div tags with class of quote
    
    quote     = sp.find('span',class_ = 'text').text[1:-1] #inside div and class we find span tag and text class to find quote
    author    = sp.find('small').text #small tag gives us author names
    author_id = sp.find('a').get('href') #anchor and href will give us author id
   
    tags = [] 
    for tag in sp.find_all('a', class_ = 'tag'): #find anchor with tag class to find all the tags 
        tags.append(tag.text)
    
    tags      = ','.join(tags)#then join the tags we found together
    
    data.append([quote,author,author_id, tags])#makes listed data function and append/add to the end of list

    
data #call data and we can see the quotes , author , id and tags in list format
```

# Scraping multiple webpages


```python
#https://quotes.toscrape.com/ has 10 pages, we need to try and automate the process
```

```python
#on the website the pages https://quotes.toscrape.com/page/2/ , page/3/ and so forth
for page in range(2,11):
    print('https://quotes.toscrape.com/'+ str(page))
    
#see that we can print each url page links
```

```python
#take some function and turn into links
data = []

for page in range(1,11):
    link = 'https://quotes.toscrape.com/page/'+ str(page)
    res  = requests.get(link) # pass link through request,and get res
    soup = BeautifulSoup(res.text, 'html.parser')
    
#take previouse automated code from above , that we made to find tags, and print them
    for sp in soup.find_all('div', class_ = 'quote'): 
    
    
        quote     = sp.find('span',class_ = 'text').text[1:-1] 
        author    = sp.find('small').text 
        author_id = sp.find('a').get('href') 

        tags = [] 
        for tag in sp.find_all('a', class_ = 'tag'):
            tags.append(tag.text)

        tags      = ','.join(tags)
    
        data.append([quote,author,author_id, tags])
```

```python
len(data)
#we check if our data is working, 100 data, 10 pages with 10 quyotes each
```
